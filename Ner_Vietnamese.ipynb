```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("NlpHUST/ner-vietnamese-electra-base")
model = AutoModelForTokenClassification.from_pretrained("NlpHUST/ner-vietnamese-electra-base")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)


```

```python
example = "Tình hình đường Hàn Thuyên  - nhà thờ Đức Bà?"
print("Câu hỏi: ", example)

ner_results = nlp(example)

# Khởi tạo danh sách để lưu tên địa điểm
locations = []
current_location = []
current_score = 0

for result in ner_results:
    if result['entity'] == 'I-LOCATION':
        # Thêm từ vào danh sách hiện tại
        current_location.append(result['word'])
        current_score = max(current_score, result['score'])  # Lưu độ tin cậy cao nhất
    else:
        # Nếu không còn thực thể địa điểm, lưu kết quả
        if current_location:
            locations.append({
                'name': ' '.join(current_location),  # Ghép các từ lại thành một cụm
                'score': current_score
            })
            # Reset danh sách
            current_location = []
            current_score = 0

# Kiểm tra nếu còn địa điểm chưa lưu
if current_location:
    locations.append({
        'name': ' '.join(current_location),
        'score': current_score
    })

# Hiển thị kết quả
for loc in locations:
    print(f"\nTên địa điểm: {loc['name']}, Độ tin cậy: {loc['score']:.4f}")
```

```python
# Đọc tệp văn bản và trích xuất tên đường
file_path = "C:\\Python\\Traffic-Prediction-Project\\test_text.txt"  # Thay bằng đường dẫn tới tệp của bạn
with open(file_path, 'r', encoding='utf-8') as file:
    sentences = file.readlines()

# Xử lý từng câu trong tệp
for i, sentence in enumerate(sentences, 1):
    print(f"\nCâu {i}: {sentence.strip()}")
    
    ner_results = nlp(sentence)
    
    # Khởi tạo danh sách để lưu tên địa điểm
    locations = []
    current_location = []
    current_score = 0

    for result in ner_results:
        if result['entity'] == 'I-LOCATION':
            # Thêm từ vào danh sách hiện tại
            current_location.append(result['word'])
            current_score = max(current_score, result['score'])  # Lưu độ tin cậy cao nhất
        else:
            # Nếu không còn thực thể địa điểm, lưu kết quả
            if current_location:
                locations.append({
                    'name': ' '.join(current_location),  # Ghép các từ lại thành một cụm
                    'score': current_score
                })
                # Reset danh sách
                current_location = []
                current_score = 0

    # Kiểm tra nếu còn địa điểm chưa lưu
    if current_location:
        locations.append({
            'name': ' '.join(current_location),
            'score': current_score
        })

    # Hiển thị kết quả cho câu hiện tại
    if locations:
        for loc in locations:
            print(f"Tên địa điểm: {loc['name']}, Độ tin cậy: {loc['score']:.4f}")
    else:
        print("Không tìm thấy địa điểm nào.")
```
